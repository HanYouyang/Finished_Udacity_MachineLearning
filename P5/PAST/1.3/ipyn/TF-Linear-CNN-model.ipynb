```python
import tensorflow as tf
sess = tf.InteractiveSession()

redataname = ['benign','malignant','radius_mean','texture_mean','perimeter_mean',
            'area_mean','smoothness_mean','compactness_mean','concavity_mean',
            'concave points_mean','symmetry_mean','fractal_dimension_mean','radius_se',
            'texture_se','perimeter_se','area_se','smoothness_se','compactness_se',
            'concavity_se','concave points_se','symmetry_se','fractal_dimension_se',
            'radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst',
            'compactness_worst','concavity_worst','concave points_worst','symmetry_worst',
            'fractal_dimension_worst']

# nddata means new data
# trndata means trianing_use new data
nddata = final_change[0:450]
trnddata = final_change[450:495]
```

```python
x = tf.placeholder("float", [None, 30])
y_ = tf.placeholder("float", [None, 2])


W = tf.Variable(tf.truncated_normal([30, 2]))
b = tf.Variable(tf.truncated_normal([2]))

y = tf.matmul(x, W) + b


cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(y, y_))

optimizer = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)

```

```python
correct_production = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy_linear = tf.reduce_mean(tf.cast(correct_production, tf.float32))

sess.run(tf.initialize_all_variables())

linear_training_accuracy  = 0

for i in range(20002):

    ntfdata = nddata.sample(n=1)
    
    x_train = np.float32(ntfdata[redataname[2:32]])
    y_train = np.float32(ntfdata[redataname[:2]])
    x_test = np.reshape(x_train, (1, 30))
    y_test = np.reshape(y_train, (1, 2))
    sess.run(optimizer, feed_dict={x:x_test, y_:y_test})
    if i%200 == 1 and i !=1:
        
        print("Step %d, linear model of TensorFlow training accuracy %f"%(i-1,  linear_training_accuracy/200))
        linear_training_accuracy  = 0
    
    else :
        linear_training_accuracy += accuracy_linear.eval(feed_dict={x:x_test, y_: y_test})
```

```python
linear_testing_accuracy  = 0
total_accrracy = 0

for i in range(3002):
    ntfdata = trnddata.sample(n=1)
    
    x_train = np.float32(ntfdata[redataname[2:32]])
    y_train = np.float32(ntfdata[redataname[:2]])
    x_test = np.reshape(x_train, (1, 30))
    y_test = np.reshape(y_train, (1, 2))
    if i%200 == 1 and i !=1:
        
        print("step %d, linear model of TensorFlow testing accuracy %f"%(i-1,  linear_testing_accuracy/200))
        total_accrracy += linear_testing_accuracy
        linear_testing_accuracy  = 0
        
    else :
        linear_testing_accuracy += accuracy_linear.eval(feed_dict={x:x_test, y_: y_test})
    
    if i%3001 == 0 and i != 0:
        print ("Average testing accuracy is %f.")%(total_accrracy / 3002)
```


## cnn

```python
def weight_varible(shape):
    initial = tf.truncated_normal(shape, stddev=0.1)
    return tf.Variable(initial)

def bias_varible(shape):
    initial = tf.constant(0.1, shape=shape)
    return tf.Variable(initial)

def conv2d(x, W):
    return tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding = 'SAME')

def max_pool_2x2(x):
    return tf.nn.max_pool(x, ksize=[1, 2, 2, 1],
                          strides=[1, 2, 2, 1], padding = 'SAME')    
```

```python
W_conv1 = weight_varible([5,5,1,30])
b_conv1 = bias_varible([30])
x_image = tf.reshape(x, [-1, 5, 6, 1])
h_conv1 = tf.nn.relu(conv2d(x_image, W_conv1)+b_conv1)
h_pool1 = max_pool_2x2(h_conv1)

W_conv2 = weight_varible([5,5,30,60])
b_conv2 = bias_varible([60])

h_conv2 = tf.nn.relu(conv2d(h_pool1, W_conv2)+b_conv2)
h_pool2 = max_pool_2x2(h_conv2)
```

```python
W_fc1 = weight_varible([2*2*60, 600])
b_fc1 = bias_varible([600])

h_pool2_flat = tf.reshape(h_pool2, [-1, 2*2*60])
h_fc1 = tf.nn.relu(tf.matmul(h_pool2_flat, W_fc1) + b_fc1)

keep_prob = tf.placeholder(tf.float32)
h_fc1_drop = tf.nn.dropout(h_fc1, keep_prob)

W_fc2 = weight_varible([600,2])
b_fc2 = bias_varible([2])
```

```python
y_conv = tf.matmul(h_fc1_drop, W_fc2) + b_fc2
#y_conv = tf.nn.softmax(tf.matmul(h_fc1_drop, W_fc2) + b_fc2)
```

```python
cross_entropy_2 = (tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(
    y_conv,  y_) +
    0.001*tf.nn.l2_loss(W_conv1) +
    0.001*tf.nn.l2_loss(W_conv2) +
    0.001*tf.nn.l2_loss(W_fc1) +
    0.001*tf.nn.l2_loss(W_fc2)))
```

```python
train_step = tf.train.AdamOptimizer(0.001).minimize(cross_entropy_2)
correct_pre_2 = tf.equal(tf.argmax(y_conv,1), tf.argmax(y_,1))
accuracy_2 = tf.reduce_mean(tf.cast(correct_pre_2, tf.float32))

sess.run(tf.initialize_all_variables())

train_accuracy = 0

for i in range(20002):
    tfdata = nddata.sample(frac=1)
    #display(tfdata.head())
    ntfdata =  tfdata[:1]
    
    x_train = np.float32(ntfdata[redataname[2:32]])
    y_train = np.float32(ntfdata[redataname[:2]])
    x_test = np.reshape(x_train, (1, 30))
    y_test = np.reshape(y_train, (1, 2))
    
    sess.run(train_step, feed_dict={x:x_test, y_: y_test, keep_prob: 0.5})
    if i%200 == 1 and i !=1:
        print("step %d, training accuracy %f"%(i-1, train_accuracy/200))
        train_accuracy = 0;
    else :
        train_accuracy += accuracy_2.eval(feed_dict={x:x_test, y_: y_test, keep_prob: 1.0})
```

```python
testing_accuracy = 0
total_testing_accuracy = 0

for i in range(3002):
    tfdata = trnddata.sample(frac=1)
    #display(tfdata.head())
    ntfdata =  tfdata[:1]
    
    x_train = np.float32(ntfdata[redataname[2:32]])
    y_train = np.float32(ntfdata[redataname[:2]])
    x_test = np.reshape(x_train, (1, 30))
    y_test = np.reshape(y_train, (1, 2))
    
    
    if i%200 == 1 and i !=1:
        print("step %d, testing accuracy %f"%(i-1, testing_accuracy/200))
        total_testing_accuracy += testing_accuracy
        testing_accuracy = 0;
    else :
        testing_accuracy += accuracy_2.eval(feed_dict={x:x_test, y_: y_test, keep_prob: 1.0})
    
    if i%3001 == 0 and i != 0:
        print ("Average CNN testing accuracy is %f.")%(total_testing_accuracy / 3002)
```
